```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = cv2.resize(gray, (10,10))

fig = plt.figure(figsize = (20,20))
ax = fig.add_subplot(111)

r =0 

for rows in gray:
    c = 0
    for pixel in rows:
        ax.annotate(pixel, xy = (r,c))
        c = c+1
    r = r + 1
ax.imshow(gray)
plt.show()
```

# edge detection using filters

```python
coin = cv2.imread('chess_board.jpg')
plt.imshow(coin)
plt.show()
# sobel x filter
sobel_x = np.array([[2,-1,0],
                   [-1,0,-1],
                    [0,-1,2]])
kernel1 = np.array([[0,-1,0],
                   [-1,4,-1],
                    [0,-1,0]])
sobel_y = sobel_x.T

filterd = cv2.filter2D(coin, -1, sobel_x)
plt.imshow(filterd)
plt.show()

filterd = cv2.filter2D(coin, -1, sobel_y)
plt.imshow(filterd)
plt.show()
```
