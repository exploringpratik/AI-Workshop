```python
import keras
import cv2
import matplotlib.pyplot as plt
import tensorflow
```

```python
# salt and pepper noise image
noise = cv2.imread('images/noise_img.png')
plt.imshow(noise, cmap='gray')
plt.show()

# using gaussian blur in the picture
blur = cv2.GaussianBlur(noise, (5,5), -1)#parameter of sigmoid function is -1 since grayscale image contains a single channel
plt.imshow(blur)
plt.show()

# using median blur in the picture to overcome gaussian blur
median = cv2.medianBlur(noise,5)
plt.imshow(median)
plt.show()
```

# Face detection

```python
face_cascade = cv2.CascadeClassifier('C:/Users/Admin/Desktop/AI workshop mpercept/resources/haarcascade_frontalface_default.xml')
# using camera to capture video
cap = cv2.VideoCapture(0)   # 0 can be replaced with video files

while True: # use camera until condition true
    ret, frame = cap.read()
#    print(ret)
   # check for valid frames
    if ret:    
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  # detects only on grayscale
        faces = face_cascade.detectMultiScale(gray,1.3,5)  #1.3 value => scaling value 5 => specify number of neighbors
        
        for (x,y,w,h) in faces:
            '''
            #inside for-loop
            img = cv2.rectangle(frame,(x,y),(x+w,y+h), (255,0,0), 2) 
            #outside for-loop
            cv2.imshow('Detecting', img)
            cv2.imwrite('me.png', img)
            k = cv2.waitKey(5) & 0xFF
            if k==27:   # check for escape key
               break
            '''
            # Gives a new window titled 'Detecting'
            cv2.rectangle(frame,(x,y),(x+w,y+h), (255,0,0), 2)    
        k = cv2.waitKey(5) & 0xFF  # converting to decimal
        cv2.imshow('Detecting', frame)
        if k==27:   # check for escape key (key=27)
            cv2.imwrite('me.png', frame)
            break
cap.release()
cv2.destroyAllWindows()
```

```python
# Face detection on images
img = cv2.imread('images/person.jpg', 1)
#plt.imshow(img)
#plt.show()

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#plt.imshow(gray)
#plt.show()

dir_ = 'C:/Users/Admin/Desktop/AI workshop mpercept/resources/haarcascade_frontalface_default.xml'
face_cascade = cv2.CascadeClassifier(dir_)

faces = face_cascade.detectMultiScale(gray,1.3,5)  #1.3 value => scaling value 5 => specify number of neighbors
faces # [bounding_box, , , ]

for (x,y,w,h) in faces:
    img = cv2.rectangle(img,(x,y),(x+w,y+h), (255,0,0), 2)
plt.imshow(img)
```
